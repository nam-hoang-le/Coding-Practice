# Implementation for one sample 

## Load data

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('data.csv', delimiter= ',').tolist()
data
```

## Data preparation

```python
def get_column(data, index): 
    result = [row[index] for row in data]
    return result 

areas = get_column(data, 0)
print(areas)
```

```python
prices = get_column(data, 1)
prices
```

```python
N = len(prices)
N
```

```python
plt.scatter(areas, prices)
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

## Training

### Prepare functions

```python
# forward 
def predict(x, w, b): 
    return x * w + b

# compute gradient 
def gradient(y_hat, y, x): 
    dw = 2 * x * (y_hat - y)
    db = 2 * (y_hat - y)
    return dw, db 

# update weights 
def update_weights(lr, db, dw, b, w): 
    w_new = w - dw * lr
    b_new = b - db * lr 
    return w_new, b_new 

# compute loss
def compute_loss(y_hat, y): 
    return (y_hat - y) ** 2 /2
```

### Initialize parameters

```python
# init weights
w = 0
b = 0
lr = 0.02 

# parameter 
epoches_max = 100
losses = []
```

### Training

```python
for epoch in range(epoches_max): 
    for i in range(N): 

        # get a sample
        data_train = prices[i]
        data_test = areas[i]

        # predict y_hat 
        y_hat = predict(data_train, w, b)

        # compute loss
        loss = compute_loss(y_hat, data_test)
        print(loss)

        # save losses 
        losses.append(loss)

        # compute gradient 
        dw, db = gradient(y_hat, data_test, data_train)

        # update weights 
        w, b = update_weights(lr, db, dw, b, w)
```

```python
print(w, b)
```

## Plot the results

```python
plt.plot(losses)
```

```python
x_data = range(2, 8)
y_data = [x * w + b for x in x_data]
plt.plot(x_data, y_data)

plt.scatter(areas, prices)
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```
